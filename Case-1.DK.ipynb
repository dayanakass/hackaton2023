```python
news = pd.read_csv('epir_train/news.csv')
news.head()
```

```python
news = news.fillna('')
```

```python
news['text'] = news['title'] + ' ' + news['body']+ ' ' + news['short_description'] 
```

```python
news['text'] = news['text'].str.strip()
```

```python
news = news[['Unnamed: 0', 'text', 'url']]
```

```python
news = news[news['text'].apply(len) > 0]
```

```python
news
```

```python
news['text'] = news['text'].str.lower()
```

```python
import re
news['text'] = news['text'].apply(lambda t: re.sub(r'[^\w\s]', '', t))
```
